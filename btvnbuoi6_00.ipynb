```python
# Step 1. Import the necessary libraries
import numpy as np 
import pandas as pd 
```

```python
# Step 2. Create a data dictionary that looks like the DataFrame below
record = [('Ivysaur', '45', 'Bulbasaur',  'yes', 'grass'),
          ('Charmeleon', '39', 'Charmander', 'no', 'fire'),
          ('Wartortle', '44', 'Squirtle', 'yes', 'water'),
          ('Metapod', '45', 'Caterpie', 'no', 'bug')]
```

```python
# Step 3. Assign it to a variable called pokemon
pokemon = pd.DataFrame(data = record, columns=['evolution', 'hp', 'name', 'pokedex', 'type'])
```

```python
pokemon 
```

```python
# Step 4. Ops...it seems the DataFrame columns are in alphabetical order. Place the order of the columns as name, type, hp, evolution, pokedex
pokemon = pokemon[['name', 'type', 'hp', 'evolution', 'pokedex']]
```

```python
pokemon
```

```python
# Step 5. Add another column called place, and insert what you have in mind.
import random
pokemon_len = len(pokemon)
birthday = [random.randrange(1954, 1990, 1) for i in range(pokemon_len)]
pokemon['birthday'] = birthday
pokemon.tail(4)
```

```python
pokemon['is_young'] = pokemon['birthday'] > 1980
pokemon.head(4)
```
